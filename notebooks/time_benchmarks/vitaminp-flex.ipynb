```python
import numpy as np
import torch
import time
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import tifffile
import tempfile
import shutil
import os
import gc
import psutil
import json

from vitaminp import VitaminPFlex
from vitaminp.inference import WSIPredictor

# ============================================
# METRIC EXTRACTION FUNCTIONS
# ============================================

def compute_segmentation_metrics(mask):
    """Compute detailed segmentation statistics"""
    unique_labels = np.unique(mask)
    n_cells = len(unique_labels) - 1  # exclude background (0)

    cell_sizes = []
    for label in unique_labels[1:]:  # skip background
        cell_sizes.append(np.sum(mask == label))

    return {
        'n_cells': n_cells,
        'mean_cell_size': np.mean(cell_sizes) if cell_sizes else 0,
        'median_cell_size': np.median(cell_sizes) if cell_sizes else 0,
        'std_cell_size': np.std(cell_sizes) if cell_sizes else 0,
        'min_cell_size': np.min(cell_sizes) if cell_sizes else 0,
        'max_cell_size': np.max(cell_sizes) if cell_sizes else 0,
    }

def measure_gpu_memory():
    """Measure current GPU memory usage"""
    if torch.cuda.is_available():
        return {
            'allocated_gb': torch.cuda.memory_allocated() / 1024**3,
            'reserved_gb': torch.cuda.memory_reserved() / 1024**3,
            'max_allocated_gb': torch.cuda.max_memory_allocated() / 1024**3,
        }
    return {'allocated_gb': 0, 'reserved_gb': 0, 'max_allocated_gb': 0}

def measure_cpu_memory():
    """Measure current CPU memory usage"""
    process = psutil.Process()
    return process.memory_info().rss / 1024**3  # GB

def load_mask_from_results(output_dir, image_shape):
    """
    Reconstruct a binary/instance mask from the output files.
    Tries parquet first (has polygon coords), falls back to geojson.
    """
    import pandas as pd

    mask = np.zeros(image_shape[:2], dtype=np.uint32)

    # Try parquet first
    parquet_files = [f for f in os.listdir(output_dir) if f.endswith('.parquet')]
    if parquet_files:
        df = pd.read_parquet(os.path.join(output_dir, parquet_files[0]))
        # Parquet from VitaminP should have polygon geometry or bbox columns
        # If it has 'geometry' as WKT or shapely objects:
        try:
            from shapely import wkt as shapely_wkt
            from shapely.geometry import shape as shapely_shape
            import shapely

            for idx, row in df.iterrows():
                label = idx + 1
                geom = row.get('geometry', None)
                if geom is None:
                    continue
                # Parse geometry if it's WKT string
                if isinstance(geom, str):
                    poly = shapely_wkt.loads(geom)
                else:
                    poly = geom

                # Rasterize polygon into mask
                minx, miny, maxx, maxy = poly.bounds
                minx, miny = max(0, int(minx)), max(0, int(miny))
                maxx, maxy = min(image_shape[1], int(maxx)+1), min(image_shape[0], int(maxy)+1)

                for y in range(miny, maxy):
                    for x in range(minx, maxx):
                        from shapely.geometry import Point
                        if poly.contains(Point(x, y)):
                            mask[y, x] = label
            return mask
        except Exception:
            pass

    # Fallback: use num_detections from results dict directly
    # and return empty mask (metrics will use n_cells from results)
    return mask


def save_temp_tif(img_array, path):
    """Save numpy array as a TIFF file for WSIPredictor"""
    tifffile.imwrite(path, img_array)


def run_vitaminp(predictor, tif_path, output_dir, run_idx=0):
    """
    Run VitaminP inference and return timing + results.
    output_dir is unique per run to avoid file conflicts.
    """
    run_output = os.path.join(output_dir, f"run_{run_idx}")
    os.makedirs(run_output, exist_ok=True)

    torch.cuda.synchronize()
    start = time.time()

    results = predictor.predict(
        wsi_path=tif_path,
        output_dir=run_output,
        branch='he_nuclei',
        filter_tissue=False,
        clean_overlaps=True,
        save_masks=False,
        save_geojson=True,
        detection_threshold=0.2,
        min_area_um=5.0,
        simplify_epsilon=None,
        coord_precision=1,
        save_parquet=True,
    )

    torch.cuda.synchronize()
    elapsed = time.time() - start

    return elapsed, results, run_output


# ============================================
# LOAD MODEL & PREDICTOR
# ============================================

print("Loading VitaminP model...")
device = 'cuda'
model = VitaminPFlex(model_size='small').to(device)
model.load_state_dict(torch.load(
    "checkpoints/vitamin_p_flex_small_fold1_best.pth",
    map_location=device
))
model.eval()

predictor = WSIPredictor(
    model=model,
    device='cuda',
    patch_size=512,
    overlap=64,
    target_mpp=0.263,
    magnification=40,
)
print("  âœ“ Model and predictor loaded")

# ============================================
# SETUP TEMP DIRECTORIES
# ============================================

TEMP_DIR = "vitaminp_benchmark_tmp"
os.makedirs(TEMP_DIR, exist_ok=True)

# Pre-save test images as TIFF (WSIPredictor needs file paths)
tile_512_png = np.array(Image.open("test_images/ovarian-he_chunk_92.png"))
tile_512_tif = os.path.join(TEMP_DIR, "tile_512.tif")
save_temp_tif(tile_512_png, tile_512_tif)

wsi_5000 = np.array(Image.open("test_images/region_5000x5000_standard.tif"))
wsi_5000_tif = os.path.join(TEMP_DIR, "wsi_5000.tif")
save_temp_tif(wsi_5000, wsi_5000_tif)

print(f"Tile shape: {tile_512_png.shape}")
print(f"WSI shape:  {wsi_5000.shape}")

# ============================================
# 1. SINGLE TILE BENCHMARK (512Ã—512)
# ============================================

print("=" * 70)
print("VITAMINP BENCHMARK 1: SINGLE TILE (512Ã—512)")
print("=" * 70)

# Reset memory
torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

# Warmup
print("Warmup runs...")
warmup_dir = os.path.join(TEMP_DIR, "warmup")
for w in range(3):
    run_vitaminp(predictor, tile_512_tif, warmup_dir, run_idx=w)

# Benchmark
n_runs = 10
times_512 = []
memory_512 = []
n_cells_512 = []

print(f"Running {n_runs} timed iterations...")
bench_dir_512 = os.path.join(TEMP_DIR, "bench_512")

for i in range(n_runs):
    torch.cuda.reset_peak_memory_stats()
    torch.cuda.empty_cache()
    gc.collect()

    elapsed, results, _ = run_vitaminp(predictor, tile_512_tif, bench_dir_512, run_idx=i)

    times_512.append(elapsed)
    memory_512.append(measure_gpu_memory()['max_allocated_gb'])
    n_cells_512.append(results['num_detections'])
    print(f"  Run {i+1}/{n_runs}: {times_512[-1]:.3f}s, {memory_512[-1]:.2f} GB, Cells: {n_cells_512[-1]}")

# Use last run's cell count as representative
n_cells_tile = n_cells_512[-1]

# We don't have a pixel mask easily, so compute cell size stats as N/A
# Use num_detections for cell count metrics
results_512 = {
    'image_size': '512Ã—512',
    'n_cells': n_cells_tile,
    'mean_time_s': np.mean(times_512),
    'std_time_s': np.std(times_512),
    'median_time_s': np.median(times_512),
    'min_time_s': np.min(times_512),
    'max_time_s': np.max(times_512),
    'mean_time_ms': np.mean(times_512) * 1000,
    'peak_memory_gb': np.mean(memory_512),
    'cells_per_second': n_cells_tile / np.mean(times_512),
    'mean_cell_size_px': 'N/A (polygon output)',
    'median_cell_size_px': 'N/A (polygon output)',
}

print("\nRESULTS:")
for key, value in results_512.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.4f}")
    else:
        print(f"  {key}: {value}")

# ============================================
# 2. WHOLE SLIDE IMAGE BENCHMARK (5000Ã—5000)
# ============================================

print("\n" + "=" * 70)
print("VITAMINP BENCHMARK 2: WHOLE SLIDE IMAGE (5000Ã—5000)")
print("=" * 70)

# Reset memory
torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

cpu_mem_before = measure_cpu_memory()
print(f"CPU memory before: {cpu_mem_before:.2f} GB")

bench_dir_wsi = os.path.join(TEMP_DIR, "bench_wsi")
elapsed_wsi, results_wsi_raw, _ = run_vitaminp(predictor, wsi_5000_tif, bench_dir_wsi, run_idx=0)

cpu_mem_after = measure_cpu_memory()
gpu_mem_wsi = measure_gpu_memory()

n_cells_wsi = results_wsi_raw['num_detections']

# Tile stats: patch_size=512, overlap=64 (same as CellSAM)
tile_size = 512
overlap_pixels = 64
stride = tile_size - overlap_pixels
n_tiles_x = int(np.ceil(wsi_5000.shape[1] / stride))
n_tiles_y = int(np.ceil(wsi_5000.shape[0] / stride))
total_tiles = n_tiles_x * n_tiles_y

results_wsi = {
    'image_size': '5000Ã—5000',
    'n_tiles': total_tiles,
    'tile_size': '512Ã—512',
    'overlap': overlap_pixels,
    'n_cells': n_cells_wsi,
    'total_time_s': elapsed_wsi,
    'total_time_min': elapsed_wsi / 60,
    'time_per_tile_s': elapsed_wsi / total_tiles,
    'time_per_tile_ms': (elapsed_wsi / total_tiles) * 1000,
    'peak_gpu_memory_gb': gpu_mem_wsi['max_allocated_gb'],
    'cpu_memory_delta_gb': cpu_mem_after - cpu_mem_before,
    'cells_per_second': n_cells_wsi / elapsed_wsi,
    'tiles_per_minute': total_tiles / (elapsed_wsi / 60),
    'throughput_mpx_per_min': (5000 * 5000 / 1e6) / (elapsed_wsi / 60),
    'cell_density_per_mm2': n_cells_wsi / ((5000 * 5000) / 1e6),
    'mean_cell_size_px': 'N/A (polygon output)',
}

print("\nRESULTS:")
for key, value in results_wsi.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.4f}")
    else:
        print(f"  {key}: {value}")

# ============================================
# 3. SCALING ANALYSIS: Different Image Sizes
# ============================================

print("\n" + "=" * 70)
print("VITAMINP BENCHMARK 3: SCALING ANALYSIS")
print("=" * 70)

test_sizes = [512, 1024, 2048, 5000]
scaling_results = []

for size in test_sizes:
    print(f"\nTesting {size}Ã—{size} image...")

    # Extract crop from center of WSI (same as CellSAM/Cellpose-SAM)
    if size <= 5000:
        center_y, center_x = wsi_5000.shape[0] // 2, wsi_5000.shape[1] // 2
        y1, y2 = center_y - size // 2, center_y + size // 2
        x1, x2 = center_x - size // 2, center_x + size // 2
        test_img = wsi_5000[y1:y2, x1:x2]
    else:
        test_img = wsi_5000

    # Save crop as temp tif
    crop_tif = os.path.join(TEMP_DIR, f"crop_{size}.tif")
    save_temp_tif(test_img, crop_tif)

    # Reset memory
    torch.cuda.reset_peak_memory_stats()
    torch.cuda.empty_cache()
    gc.collect()

    # Run
    bench_dir_scale = os.path.join(TEMP_DIR, f"bench_scale_{size}")
    elapsed, results_scale, _ = run_vitaminp(predictor, crop_tif, bench_dir_scale, run_idx=0)

    mem = measure_gpu_memory()
    n_cells_scale = results_scale['num_detections']

    scaling_results.append({
        'size': size,
        'n_pixels': size * size,
        'n_cells': n_cells_scale,
        'time_s': elapsed,
        'time_ms': elapsed * 1000,
        'memory_gb': mem['max_allocated_gb'],
        'cells_per_sec': n_cells_scale / elapsed if elapsed > 0 else 0,
        'pixels_per_sec': (size * size) / elapsed if elapsed > 0 else 0,
    })

    print(f"  Time: {elapsed:.2f}s, Cells: {n_cells_scale}, Memory: {mem['max_allocated_gb']:.2f} GB")

# ============================================
# 4. COMPILE ALL RESULTS
# ============================================

print("\n" + "=" * 70)
print("VITAMINP COMPREHENSIVE BENCHMARK SUMMARY")
print("=" * 70)

summary_df = pd.DataFrame([
    {
        'Test': 'Single Tile (512Ã—512)',
        'Size': '512Ã—512',
        'Cells': results_512['n_cells'],
        'Time (s)': results_512['mean_time_s'],
        'Time (ms)': results_512['mean_time_ms'],
        'Memory (GB)': results_512['peak_memory_gb'],
        'Cells/sec': results_512['cells_per_second'],
    },
    {
        'Test': 'WSI (5000Ã—5000)',
        'Size': '5000Ã—5000',
        'Cells': results_wsi['n_cells'],
        'Time (s)': results_wsi['total_time_s'],
        'Time (ms)': results_wsi['time_per_tile_ms'],
        'Memory (GB)': results_wsi['peak_gpu_memory_gb'],
        'Cells/sec': results_wsi['cells_per_second'],
    }
])

print("\n" + summary_df.to_string(index=False))

scaling_df = pd.DataFrame(scaling_results)
print("\n" + "=" * 70)
print("SCALING ANALYSIS")
print("=" * 70)
print(scaling_df.to_string(index=False))

# ============================================
# 5. SAVE ALL RESULTS
# ============================================

summary_df.to_csv('vitaminp_benchmark_summary.csv', index=False)
scaling_df.to_csv('vitaminp_benchmark_scaling.csv', index=False)

# For JSON, convert any non-serializable values
def make_serializable(obj):
    if isinstance(obj, (np.integer,)):
        return int(obj)
    if isinstance(obj, (np.floating,)):
        return float(obj)
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    return obj

all_results = {
    'single_tile_512': {k: make_serializable(v) for k, v in results_512.items()},
    'wsi_5000': {k: make_serializable(v) for k, v in results_wsi.items()},
    'scaling': [{k: make_serializable(v) for k, v in r.items()} for r in scaling_results],
    'hardware': {
        'gpu': torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'N/A',
        'gpu_memory_total_gb': torch.cuda.get_device_properties(0).total_memory / 1024 ** 3 if torch.cuda.is_available() else 0,
    },
    'benchmark_date': time.strftime('%Y-%m-%d %H:%M:%S'),
}

with open('vitaminp_benchmark_complete.json', 'w') as f:
    json.dump(all_results, f, indent=2)

print("\nâœ“ Results saved:")
print("  - vitaminp_benchmark_summary.csv")
print("  - vitaminp_benchmark_scaling.csv")
print("  - vitaminp_benchmark_complete.json")

# ============================================
# 6. VISUALIZATION
# ============================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Scaling - Time vs Image Size
axes[0, 0].plot(scaling_df['size'], scaling_df['time_s'], 'o-', linewidth=2, markersize=8)
axes[0, 0].set_xlabel('Image Size (pixels)', fontsize=12)
axes[0, 0].set_ylabel('Time (seconds)', fontsize=12)
axes[0, 0].set_title('Processing Time vs Image Size', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].set_xscale('log')
axes[0, 0].set_yscale('log')

# Plot 2: Scaling - Memory vs Image Size
axes[0, 1].plot(scaling_df['size'], scaling_df['memory_gb'], 's-', color='red', linewidth=2, markersize=8)
axes[0, 1].set_xlabel('Image Size (pixels)', fontsize=12)
axes[0, 1].set_ylabel('Peak GPU Memory (GB)', fontsize=12)
axes[0, 1].set_title('Memory Usage vs Image Size', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_xscale('log')

# Plot 3: Throughput
axes[1, 0].bar(['512Ã—512\n(single)', '5000Ã—5000\n(WSI)'],
               [results_512['cells_per_second'], results_wsi['cells_per_second']],
               color=['steelblue', 'coral'])
axes[1, 0].set_ylabel('Cells/Second', fontsize=12)
axes[1, 0].set_title('Segmentation Throughput', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Plot 4: Time per tile comparison
labels = ['Single Tile\n(512Ã—512)', 'WSI Tile Average\n(5000Ã—5000)']
times_ms = [results_512['mean_time_ms'], results_wsi['time_per_tile_ms']]
axes[1, 1].bar(labels, times_ms, color=['steelblue', 'coral'])
axes[1, 1].set_ylabel('Time (milliseconds)', fontsize=12)
axes[1, 1].set_title('Time per Tile Comparison', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('vitaminp_benchmark_plots.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ Visualization saved to vitaminp_benchmark_plots.png")

# ============================================
# CLEANUP TEMP FILES
# ============================================
shutil.rmtree(TEMP_DIR, ignore_errors=True)
print(f"âœ“ Cleaned up temp directory: {TEMP_DIR}")

print("\n" + "=" * 70)
print("BENCHMARK COMPLETE!")
print("=" * 70)
print("\nKey Metrics for Manuscript:")
print(f"  â€¢ Time per 512Ã—512 tile: {results_512['mean_time_ms']:.1f} ms")
print(f"  â€¢ Peak memory (single tile): {results_512['peak_memory_gb']:.2f} GB")
print(f"  â€¢ WSI processing time (5000Ã—5000): {results_wsi['total_time_min']:.2f} minutes")
print(f"  â€¢ Throughput: {results_wsi['cells_per_second']:.1f} cells/second")
print(f"  â€¢ Tiles per minute: {results_wsi['tiles_per_minute']:.1f}")
print("=" * 70)
```
