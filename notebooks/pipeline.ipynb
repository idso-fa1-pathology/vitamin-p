```python
# Cell 10: Train with the trainer
import torch
from vitamin_p import VitaminPFlex, VitaminPTrainer

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Initialize model
model = VitaminPFlex(model_size='large')

# Initialize trainer
trainer = VitaminPTrainer(
    model=model,
    train_loader=train_loader,
    val_loader=val_loader,
    device=device,
    lr=1e-4,
    weight_decay=1e-4,
    use_wandb=False,  # Set to True to enable W&B
    project_name="vitamin-p-test",
    run_name="flex-base-test",
    checkpoint_dir="checkpoints"
)

# Train for 2 epochs (test)
trainer.train(epochs=2, use_augmentations=True)
```
