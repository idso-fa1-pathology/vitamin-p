## vitamin-p Flex Twice

```python
import numpy as np
import torch
import time
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import tifffile
import tempfile
import shutil
import os
import gc
import psutil
import json
import tifffile

# Ensure we are using the new code
from vitaminp import VitaminPFlex
from vitaminp.inference import WSIPredictor

# ============================================
# CONFIGURATION
# ============================================

BATCH_SIZE = 128        # <--- ADJUST THIS if you get CUDA OOM (try 32 or 16)
OUTPUT_DIR = "benchmark_results/vitaminp_optimized"
os.makedirs(OUTPUT_DIR, exist_ok=True)

TEMP_DIR = os.path.join(OUTPUT_DIR, "temp")
os.makedirs(TEMP_DIR, exist_ok=True)

# ============================================
# METRIC EXTRACTION FUNCTIONS
# ============================================

def measure_gpu_memory():
    """Measure current GPU memory usage"""
    if torch.cuda.is_available():
        return {
            'allocated_gb': torch.cuda.memory_allocated() / 1024**3,
            'reserved_gb': torch.cuda.memory_reserved() / 1024**3,
            'max_allocated_gb': torch.cuda.max_memory_allocated() / 1024**3,
        }
    return {'allocated_gb': 0, 'reserved_gb': 0, 'max_allocated_gb': 0}

def measure_cpu_memory():
    """Measure current CPU memory usage"""
    process = psutil.Process()
    return process.memory_info().rss / 1024**3  # GB

def save_temp_tif(img_array, path):
    """Save numpy array as a TIFF file for WSIPredictor"""
    tifffile.imwrite(path, img_array)

def run_vitaminp_single(predictor, tif_path, output_dir, branch, run_idx=0):
    """Run VitaminP inference for single branch (nuclei OR cells)"""
    run_output = os.path.join(output_dir, f"run_{run_idx}")
    os.makedirs(run_output, exist_ok=True)

    torch.cuda.synchronize()
    start = time.time()

    results = predictor.predict(
        wsi_path=tif_path,
        output_dir=run_output,
        branches=[branch],
        filter_tissue=False,
        clean_overlaps=True,
        save_masks=False,
        save_geojson=True,
        detection_threshold=0.5,
        min_area_um=10.0,
        simplify_epsilon=None,
        coord_precision=1,
        save_parquet=True,
    )

    torch.cuda.synchronize()
    elapsed = time.time() - start

    return elapsed, results, run_output

def run_vitaminp_dual(predictor, tif_path, output_dir, run_idx=0):
    """Run VitaminP inference for BOTH branches (nuclei AND cells)"""
    run_output = os.path.join(output_dir, f"run_{run_idx}")
    os.makedirs(run_output, exist_ok=True)

    torch.cuda.synchronize()
    start = time.time()

    # The predictor will now automatically batch these requests together!
    results = predictor.predict(
        wsi_path=tif_path,
        output_dir=run_output,
        branches=['he_nuclei', 'he_cell'],  # BOTH!
        filter_tissue=False,
        clean_overlaps=True,
        save_masks=False,
        save_geojson=True,
        detection_threshold=0.2,
        min_area_um=5.0,
        simplify_epsilon=None,
        coord_precision=1,
        save_parquet=False,
    )

    torch.cuda.synchronize()
    elapsed = time.time() - start

    return elapsed, results, run_output

# ============================================
# LOAD MODEL & PREDICTOR
# ============================================

print("Loading VitaminP model...")
device = 'cuda'
model = VitaminPFlex(model_size='base').to(device)
model.load_state_dict(torch.load(
    "checkpoints/vitamin_p_flex_base_fold21_best.pth",
    map_location=device
))
model.eval()

print(f"Initializing Predictor with Batch Size: {BATCH_SIZE}")
predictor = WSIPredictor(
    model=model,
    device='cuda',
    patch_size=512,
    overlap=64,
    target_mpp=0.263,
    magnification=40,
    batch_size=BATCH_SIZE  # <--- The Optimisation!
)
print("  ‚úì Model and predictor loaded")

# ============================================
# SETUP TEMP TEST IMAGES
# ============================================

# Pre-save test images as TIFF
tile_512_png = np.array(Image.open("test_images/ovarian-he_chunk_92.png"))
tile_512_tif = os.path.join(TEMP_DIR, "tile_512.tif")
save_temp_tif(tile_512_png, tile_512_tif)

wsi_5000 = tifffile.imread("test_images/breat_cancer_15000x15000.tiff")
wsi_5000_tif = os.path.join(TEMP_DIR, "wsi_5000.tif")
save_temp_tif(wsi_5000, wsi_5000_tif)

print(f"Tile shape: {tile_512_png.shape}")
print(f"WSI shape:  {wsi_5000.shape}")

# ============================================
# 1. SINGLE-TASK BENCHMARK: Nuclei Only
# ============================================

print("=" * 70)
print("VITAMINP BENCHMARK 1: SINGLE-TASK (Nuclei Only, 512√ó512)")
print("=" * 70)
print("Note: Batching overhead might make single-tile latency slightly higher.")
print("The real speedup will show in the WSI benchmarks (Steps 4 & 5).")

torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

# Warmup
print("Warmup runs...")
warmup_dir = os.path.join(TEMP_DIR, "warmup_nuclei")
for w in range(3):
    run_vitaminp_single(predictor, tile_512_tif, warmup_dir, 'he_nuclei', run_idx=w)

# Benchmark
n_runs = 10
times_nuclei = []
memory_nuclei = []
n_nuclei_list = []

print(f"Running {n_runs} timed iterations...")
bench_dir = os.path.join(TEMP_DIR, "bench_nuclei")

for i in range(n_runs):
    torch.cuda.reset_peak_memory_stats()
    torch.cuda.empty_cache()
    gc.collect()

    elapsed, results, _ = run_vitaminp_single(predictor, tile_512_tif, bench_dir, 'he_nuclei', run_idx=i)

    times_nuclei.append(elapsed)
    memory_nuclei.append(measure_gpu_memory()['max_allocated_gb'])
    
    # FIX: Handle nested result structure
    count = results.get('he_nuclei', {}).get('num_detections', 0) if 'he_nuclei' in results else results.get('num_detections', 0)
    
    n_nuclei_list.append(count)
    print(f"  Run {i+1}/{n_runs}: {times_nuclei[-1]:.3f}s, {memory_nuclei[-1]:.2f} GB, Nuclei: {n_nuclei_list[-1]}")

results_nuclei = {
    'task': 'Nuclei Only',
    'image_size': '512√ó512',
    'n_instances': n_nuclei_list[-1],
    'mean_time_s': np.mean(times_nuclei),
    'std_time_s': np.std(times_nuclei),
    'mean_time_ms': np.mean(times_nuclei) * 1000,
    'peak_memory_gb': np.mean(memory_nuclei),
    'instances_per_second': n_nuclei_list[-1] / np.mean(times_nuclei),
}

# ============================================
# 2. SINGLE-TASK BENCHMARK: Cells Only
# ============================================

print("\n" + "=" * 70)
print("VITAMINP BENCHMARK 2: SINGLE-TASK (Cells Only, 512√ó512)")
print("=" * 70)

torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

# Warmup
print("Warmup runs...")
warmup_dir = os.path.join(TEMP_DIR, "warmup_cells")
for w in range(3):
    run_vitaminp_single(predictor, tile_512_tif, warmup_dir, 'he_cell', run_idx=w)

# Benchmark
times_cells = []
memory_cells = []
n_cells_list = []

print(f"Running {n_runs} timed iterations...")
bench_dir = os.path.join(TEMP_DIR, "bench_cells")

for i in range(n_runs):
    torch.cuda.reset_peak_memory_stats()
    torch.cuda.empty_cache()
    gc.collect()

    elapsed, results, _ = run_vitaminp_single(predictor, tile_512_tif, bench_dir, 'he_cell', run_idx=i)

    times_cells.append(elapsed)
    memory_cells.append(measure_gpu_memory()['max_allocated_gb'])
    
    # FIX: Handle nested result structure
    count = results.get('he_cell', {}).get('num_detections', 0) if 'he_cell' in results else results.get('num_detections', 0)
    
    n_cells_list.append(count)
    print(f"  Run {i+1}/{n_runs}: {times_cells[-1]:.3f}s, {memory_cells[-1]:.2f} GB, Cells: {n_cells_list[-1]}")

results_cells = {
    'task': 'Cells Only',
    'image_size': '512√ó512',
    'n_instances': n_cells_list[-1],
    'mean_time_s': np.mean(times_cells),
    'std_time_s': np.std(times_cells),
    'mean_time_ms': np.mean(times_cells) * 1000,
    'peak_memory_gb': np.mean(memory_cells),
    'instances_per_second': n_cells_list[-1] / np.mean(times_cells),
}

# ============================================
# 3. DUAL-TASK BENCHMARK: Nuclei + Cells
# ============================================

print("\n" + "=" * 70)
print("‚≠ê VITAMINP BENCHMARK 3: DUAL-TASK (Nuclei + Cells, 512√ó512)")
print("=" * 70)

torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

# Warmup
print("Warmup runs...")
warmup_dir = os.path.join(TEMP_DIR, "warmup_dual")
for w in range(3):
    run_vitaminp_dual(predictor, tile_512_tif, warmup_dir, run_idx=w)

# Benchmark
times_dual = []
memory_dual = []
n_nuclei_dual = []
n_cells_dual = []

print(f"Running {n_runs} timed iterations...")
bench_dir = os.path.join(TEMP_DIR, "bench_dual")

for i in range(n_runs):
    torch.cuda.reset_peak_memory_stats()
    torch.cuda.empty_cache()
    gc.collect()

    elapsed, results, _ = run_vitaminp_dual(predictor, tile_512_tif, bench_dir, run_idx=i)

    times_dual.append(elapsed)
    memory_dual.append(measure_gpu_memory()['max_allocated_gb'])
    
    # Extract counts (Safe Extraction)
    nuclei_count = results.get('he_nuclei', {}).get('num_detections', 0)
    cells_count = results.get('he_cell', {}).get('num_detections', 0)
    
    n_nuclei_dual.append(nuclei_count)
    n_cells_dual.append(cells_count)
    
    total_instances = n_nuclei_dual[-1] + n_cells_dual[-1]
    print(f"  Run {i+1}/{n_runs}: {times_dual[-1]:.3f}s, {memory_dual[-1]:.2f} GB, Nuclei: {n_nuclei_dual[-1]}, Cells: {n_cells_dual[-1]}")

total_instances_dual = n_nuclei_dual[-1] + n_cells_dual[-1]
mean_time_dual = np.mean(times_dual)
mean_time_single = np.mean(times_nuclei)

results_dual = {
    'task': 'Nuclei + Cells (Dual)',
    'image_size': '512√ó512',
    'n_nuclei': n_nuclei_dual[-1],
    'n_cells': n_cells_dual[-1],
    'total_instances': total_instances_dual,
    'mean_time_s': mean_time_dual,
    'std_time_s': np.std(times_dual),
    'mean_time_ms': mean_time_dual * 1000,
    'peak_memory_gb': np.mean(memory_dual),
    'instances_per_second': total_instances_dual / mean_time_dual if mean_time_dual > 0 else 0,
    'overhead_vs_single_pct': ((mean_time_dual / mean_time_single) - 1) * 100 if mean_time_single > 0 else 0,
}

# ============================================
# 4. WSI BENCHMARK: Nuclei Only
# ============================================

print("\n" + "=" * 70)
print("VITAMINP BENCHMARK 4: WSI SINGLE-TASK (Nuclei Only, 5000√ó5000)")
print("=" * 70)
print("üöÄ This is where BATCHING speedup should happen!")

torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

bench_dir_wsi = os.path.join(TEMP_DIR, "bench_wsi_nuclei")
elapsed_wsi, results_wsi_raw, _ = run_vitaminp_single(predictor, wsi_5000_tif, bench_dir_wsi, 'he_nuclei', run_idx=0)

gpu_mem_wsi = measure_gpu_memory()

# FIX: Nested Extraction
n_nuclei_wsi = results_wsi_raw.get('he_nuclei', {}).get('num_detections', 0)

tile_size = 512
overlap_pixels = 64
stride = tile_size - overlap_pixels
n_tiles_x = int(np.ceil(wsi_5000.shape[1] / stride))
n_tiles_y = int(np.ceil(wsi_5000.shape[0] / stride))
total_tiles = n_tiles_x * n_tiles_y

results_wsi_nuclei = {
    'task': 'Nuclei Only',
    'image_size': '5000√ó5000',
    'n_tiles': total_tiles,
    'n_instances': n_nuclei_wsi,
    'total_time_s': elapsed_wsi,
    'total_time_min': elapsed_wsi / 60,
    'time_per_tile_ms': (elapsed_wsi / total_tiles) * 1000,
    'peak_gpu_memory_gb': gpu_mem_wsi['max_allocated_gb'],
    'instances_per_second': n_nuclei_wsi / elapsed_wsi,
    'throughput_mpx_per_min': (5000 * 5000 / 1e6) / (elapsed_wsi / 60),
}

print("\nRESULTS:")
for key, value in results_wsi_nuclei.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.4f}")
    else:
        print(f"  {key}: {value}")

# ============================================
# 5. WSI BENCHMARK: Dual-Task
# ============================================

print("\n" + "=" * 70)
print("‚≠ê VITAMINP BENCHMARK 5: WSI DUAL-TASK (Nuclei + Cells, 5000√ó5000)")
print("=" * 70)
print("üöÄ Checking throughput with Joint Constrained Watershed + Batching...")

torch.cuda.reset_peak_memory_stats()
torch.cuda.empty_cache()
gc.collect()

bench_dir_wsi_dual = os.path.join(TEMP_DIR, "bench_wsi_dual")
elapsed_wsi_dual, results_wsi_dual_raw, _ = run_vitaminp_dual(predictor, wsi_5000_tif, bench_dir_wsi_dual, run_idx=0)

gpu_mem_wsi_dual = measure_gpu_memory()

# Extract counts
nuclei_count = results_wsi_dual_raw.get('he_nuclei', {}).get('num_detections', 0)
cells_count = results_wsi_dual_raw.get('he_cell', {}).get('num_detections', 0)
total_instances_wsi = nuclei_count + cells_count

results_wsi_dual = {
    'task': 'Nuclei + Cells (Dual)',
    'image_size': '5000√ó5000',
    'n_tiles': total_tiles,
    'n_nuclei': nuclei_count,
    'n_cells': cells_count,
    'total_instances': total_instances_wsi,
    'total_time_s': elapsed_wsi_dual,
    'total_time_min': elapsed_wsi_dual / 60,
    'time_per_tile_ms': (elapsed_wsi_dual / total_tiles) * 1000,
    'peak_gpu_memory_gb': gpu_mem_wsi_dual['max_allocated_gb'],
    'instances_per_second': total_instances_wsi / elapsed_wsi_dual if elapsed_wsi_dual > 0 else 0,
    'throughput_mpx_per_min': (5000 * 5000 / 1e6) / (elapsed_wsi_dual / 60) if elapsed_wsi_dual > 0 else 0,
    'overhead_vs_single_pct': ((elapsed_wsi_dual / elapsed_wsi) - 1) * 100 if elapsed_wsi > 0 else 0,
}

print("\nRESULTS:")
for key, value in results_wsi_dual.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.4f}")
    else:
        print(f"  {key}: {value}")

print("\n‚≠ê KEY INSIGHTS:")
print(f"  Single-task WSI: {elapsed_wsi / 60:.2f} min")
print(f"  Dual-task WSI: {elapsed_wsi_dual / 60:.2f} min")
print(f"  If running sequentially: ~{(elapsed_wsi * 2) / 60:.2f} min")
print(f"  üöÄ Dual-task speedup: {(elapsed_wsi * 2) / elapsed_wsi_dual:.2f}√ó")

# ============================================
# 6. COMPILE ALL RESULTS
# ============================================

print("\n" + "=" * 70)
print("VITAMINP COMPREHENSIVE BENCHMARK SUMMARY")
print("=" * 70)

summary_df = pd.DataFrame([
    {
        'Test': 'Single Tile - Nuclei',
        'Task': 'Nuclei',
        'Size': '512√ó512',
        'Instances': results_nuclei['n_instances'],
        'Time (s)': results_nuclei['mean_time_s'],
        'Memory (GB)': results_nuclei['peak_memory_gb'],
        'Inst/sec': results_nuclei['instances_per_second'],
    },
    {
        'Test': 'Single Tile - Cells',
        'Task': 'Cells',
        'Size': '512√ó512',
        'Instances': results_cells['n_instances'],
        'Time (s)': results_cells['mean_time_s'],
        'Memory (GB)': results_cells['peak_memory_gb'],
        'Inst/sec': results_cells['instances_per_second'],
    },
    {
        'Test': 'Single Tile - DUAL',
        'Task': 'Both',
        'Size': '512√ó512',
        'Instances': results_dual['total_instances'],
        'Time (s)': results_dual['mean_time_s'],
        'Memory (GB)': results_dual['peak_memory_gb'],
        'Inst/sec': results_dual['instances_per_second'],
    },
    {
        'Test': 'WSI - Nuclei',
        'Task': 'Nuclei',
        'Size': '5000√ó5000',
        'Instances': results_wsi_nuclei['n_instances'],
        'Time (s)': results_wsi_nuclei['total_time_s'],
        'Memory (GB)': results_wsi_nuclei['peak_gpu_memory_gb'],
        'Inst/sec': results_wsi_nuclei['instances_per_second'],
    },
    {
        'Test': 'WSI - DUAL',
        'Task': 'Both',
        'Size': '5000√ó5000',
        'Instances': results_wsi_dual['total_instances'],
        'Time (s)': results_wsi_dual['total_time_s'],
        'Memory (GB)': results_wsi_dual['peak_gpu_memory_gb'],
        'Inst/sec': results_wsi_dual['instances_per_second'],
    }
])

print("\n" + summary_df.to_string(index=False))

# ============================================
# 7. SAVE ALL RESULTS
# ============================================

summary_df.to_csv(os.path.join(OUTPUT_DIR, 'vitaminp_benchmark_summary.csv'), index=False)

def make_serializable(obj):
    if isinstance(obj, (np.integer,)):
        return int(obj)
    if isinstance(obj, (np.floating,)):
        return float(obj)
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    return obj

all_results = {
    'single_tile_nuclei': {k: make_serializable(v) for k, v in results_nuclei.items()},
    'single_tile_cells': {k: make_serializable(v) for k, v in results_cells.items()},
    'single_tile_dual': {k: make_serializable(v) for k, v in results_dual.items()},
    'wsi_nuclei': {k: make_serializable(v) for k, v in results_wsi_nuclei.items()},
    'wsi_dual': {k: make_serializable(v) for k, v in results_wsi_dual.items()},
    'hardware': {
        'gpu': torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'N/A',
        'gpu_memory_total_gb': torch.cuda.get_device_properties(0).total_memory / 1024**3 if torch.cuda.is_available() else 0,
        'batch_size': BATCH_SIZE
    },
    'benchmark_date': time.strftime('%Y-%m-%d %H:%M:%S'),
}

with open(os.path.join(OUTPUT_DIR, 'vitaminp_benchmark_complete.json'), 'w') as f:
    json.dump(all_results, f, indent=2)

print("\n‚úì Results saved to:", OUTPUT_DIR)
print("  - vitaminp_benchmark_summary.csv")
print("  - vitaminp_benchmark_complete.json")

# ============================================
# 8. VISUALIZATION
# ============================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Single Tile Comparison
tasks = ['Nuclei\nOnly', 'Cells\nOnly', 'Both\n(Dual)']
times = [results_nuclei['mean_time_ms'], results_cells['mean_time_ms'], results_dual['mean_time_ms']]
axes[0, 0].bar(tasks, times, color=['steelblue', 'coral', 'green'])
axes[0, 0].set_ylabel('Time (ms)', fontsize=12)
axes[0, 0].set_title('Single Tile Latency (512√ó512)', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3, axis='y')

# Plot 2: WSI Comparison
wsi_tasks = ['Nuclei\nOnly', 'Both\n(Dual)', 'Sequential\n(2√ó Single)']
wsi_times = [
    results_wsi_nuclei['total_time_min'],
    results_wsi_dual['total_time_min'],
    results_wsi_nuclei['total_time_min'] * 2
]
colors = ['steelblue', 'green', 'red']
axes[0, 1].bar(wsi_tasks, wsi_times, color=colors)
axes[0, 1].set_ylabel('Time (minutes)', fontsize=12)
axes[0, 1].set_title('WSI Throughput (5000√ó5000)', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3, axis='y')

# Plot 3: Throughput
throughputs = [
    results_nuclei['instances_per_second'],
    results_cells['instances_per_second'],
    results_dual['instances_per_second']
]
axes[1, 0].bar(tasks, throughputs, color=['steelblue', 'coral', 'green'])
axes[1, 0].set_ylabel('Instances/Second', fontsize=12)
axes[1, 0].set_title('Segmentation Throughput', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Plot 4: Memory Usage
memory_vals = [
    results_nuclei['peak_memory_gb'],
    results_cells['peak_memory_gb'],
    results_dual['peak_memory_gb']
]
axes[1, 1].bar(tasks, memory_vals, color=['steelblue', 'coral', 'green'])
axes[1, 1].set_ylabel('Peak GPU Memory (GB)', fontsize=12)
axes[1, 1].set_title('Memory Usage (Peak)', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'vitaminp_benchmark_plots.png'), dpi=300, bbox_inches='tight')
print("  - vitaminp_benchmark_plots.png")
plt.close()

# ============================================
# CLEANUP
# ============================================

print("\n" + "=" * 70)
print("BENCHMARK COMPLETE!")
print("=" * 70)
print("\n‚≠ê KEY FINDINGS:")
print(f"  Single Tile - Nuclei: {results_nuclei['mean_time_ms']:.1f} ms")
print(f"  Single Tile - Dual: {results_dual['mean_time_ms']:.1f} ms (+{results_dual['overhead_vs_single_pct']:.1f}%)")
print(f"  WSI - Nuclei: {results_wsi_nuclei['total_time_min']:.2f} min")
print(f"  WSI - Dual: {results_wsi_dual['total_time_min']:.2f} min")
print(f"  üöÄ Dual-task speedup vs sequential: {(results_wsi_nuclei['total_time_s'] * 2) / results_wsi_dual['total_time_s']:.2f}√ó")
print("=" * 70)
```
